```python
from src.data import get_dataloader
import torchvision.utils as vutils
import matplotlib.pyplot as plt

loader = get_dataloader(batch_size=64, image_size=64)
real_batch = next(iter(loader))[0]
print("Batch shape:", real_batch.shape)
assert real_batch.shape[1:] == (3, 64, 64), "Unexpected image shape"
print("Value range:", real_batch.min().item(), "to", real_batch.max().item())
```

```python
grid = vutils.make_grid(real_batch, padding=2, normalize=True)
plt.figure(figsize=(8, 8))
plt.axis("off")
plt.title("Real Anime Faces — Sample Batch")
plt.imshow(grid.permute(1, 2, 0))
plt.show()
```
