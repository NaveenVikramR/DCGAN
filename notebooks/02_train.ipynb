## Check the progression and loss curve

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.figure(figsize=(6, 6))
plt.imshow(mpimg.imread(f"{RESULTS_DIR}/loss_curve.png"))
plt.axis("off")
plt.show()
```

```python
# Sanity-check the noise -> face progression across epochs
import glob

samples = sorted(glob.glob(f"{RESULTS_DIR}/samples/epoch_*.png"))
print(f"{len(samples)} epoch sample grids found")
for path in [samples[0], samples[len(samples) // 2], samples[-1]]:
    plt.figure(figsize=(6, 6))
    plt.title(path.split("/")[-1])
    plt.imshow(mpimg.imread(path))
    plt.axis("off")
    plt.show()
```
